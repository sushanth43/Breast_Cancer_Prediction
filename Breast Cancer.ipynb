```python
data = data.drop(["id","Unnamed: 32"],axis=1)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
y = data['diagnosis']
x = data.drop(['diagnosis'],axis=1)
```

```python
x.corr()
```

```python
col = ["perimeter_mean","area_mean","concave points_mean","radius_worst",
       "area_worst","perimeter_worst"]
x = x.drop(col,axis=1)
```

```python
x.corr()
```

```python
c = x.corr()
filter1 = c['texture_mean'] >= 0.8
filter2 = c['texture_mean'] <= -0.8
c[filter1 | filter2].index
```

```python
c = x.corr()
filter1 = c['smoothness_mean'] >= 0.8
filter2 = c['smoothness_mean'] <= -0.8
c[filter1 | filter2].index
```

```python
x.describe()
```

```python
col = ['radius_mean', 'texture_mean', 'smoothness_mean', 'compactness_mean',
       'symmetry_mean', 'fractal_dimension_mean', 'radius_se', 'texture_se',
       'smoothness_se', 'compactness_se', 'concave points_se', 'symmetry_se',
       'symmetry_worst', 'fractal_dimension_worst']
len(col)
```

```python
 x_new=x[col]
```

```python
x_new.info()
```

```python
x_new=x_new.values
```

```python
x_new.shape
```

```python
y.shape
```

```python
y=y.values
```

```python
### Standard Scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_scaled = ss.fit_transform(x_new)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x_scaled,y,train_size=0.80,random_state=1)
```

```python
import tensorflow.keras as tf
```

```python
### Build an empty model
model = tf.models.Sequential()
### Adding the layers
model.add(tf.layers.Dense(16,input_dim=14,activation='relu'))  ## adding a i/p layer and a hidden layer
model.add(tf.layers.Dense(32,activation='relu'))   ## 2nd hidden layer
model.add(tf.layers.Dense(48,activation='relu'))  
model.add(tf.layers.Dense(1,activation = "sigmoid"))
### Compile model
model.compile(loss='binary_crossentropy',optimizer="adam",metrics=['accuracy'])
```

```python
history = model.fit(xtrain,ytrain,epochs = 500)
```

```python
### Prediction
ypred = model.predict(xtest)
ypred = ypred.round()
ypred = ypred.reshape(len(ypred))
ypred
```

```python
ytest
```

```python
model.evaluate(xtest,ytest)
```

```python
count = 0
for i in range(len(ytest)):
    if ytest[i] == ypred[i]:
        count = count+1
count
```
